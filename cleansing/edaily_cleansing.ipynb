## 클렌징

```python
df = pd.read_csv(r'C:\Users\ran14\Workspace\BOK_workspace\crawl_edaily\edaily(20140811, 20240811).csv')
pd.set_option('display.max_colwidth', None)
df
```

```python
def cleansing_text(text):
    # 종목 코드 (괄호 안 6자리 숫자) 제거
    text = re.sub(r'\(\d{6}\)', '', text)
    
    # 특수 기호 및 광고성 텍스트 제거
    text = re.sub(r'▶ 관련기사 ◀', '', text)
    text = re.sub(r'☞[^☞]*', '', text)
    
    # 맥락과 관련 없는 리스트 및 간추린 소식 제거
    text = re.sub(r'\d+\.\s[^<]*', '', text)
    text = re.sub(r'<간추린 소식>[^-]*', '', text)
    
    # 구두점, 따옴표, 기타 특수 문자 제거
    text = re.sub(r'[▲△▶▼▽◆◇=ㆍ/·.,;:!?\'"‘’“”~&]', '', text)
    
    # '사진', '표'와 같은 텍스트 제거
    text = re.sub(r'사진|표', '', text)
    
    # 필요한 공백 제거
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
# TITLE과 CONTENTS 컬럼에만 cleansing_text 함수 적용
df['TITLE'] = df['TITLE'].apply(cleansing_text)
df['CONTENTS'] = df['CONTENTS'].apply(cleansing_text)

df_cleaned = df[['TITLE', 'CONTENTS', 'URL']]
df_cleaned
```

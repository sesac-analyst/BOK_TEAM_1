```python
import os
import csv

# Directory containing the article files
directory = r"C:\Users\rapha\Desktop\workspace\study\PJT\crawling"
# Output file for the combined CSV
output_csv = r"C:\Users\rapha\Desktop\workspace\study\PJT\crawling\combined_output.csv"

# Function to read the title, date, and content from a file
def extract_info(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        title_index = lines.index('제목\n') + 1 if '제목\n' in lines else 0
        title = lines[title_index].strip()
        date_index = '_'.join(os.path.basename(file_path).split('_')[:3])  # Extract the date from the file name
        content = ''.join(lines[title_index + 1:]).strip()
    return date_index, title, content

# Write to the combined CSV file
with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Date', 'Title', 'Content'])  # Write header
    
    # Process each file in the directory
    for i, filename in enumerate(os.listdir(directory), start=1):
        if filename.endswith(".text"):  # Check if it's a text file
            file_path = os.path.join(directory, filename)
            date, title, content = extract_info(file_path)
            csvwriter.writerow([date, title, content])  # Write data

print("Conversion to a single CSV file completed.")
```

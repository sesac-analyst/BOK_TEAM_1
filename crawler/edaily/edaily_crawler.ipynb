```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

## 개별 크롤링 코드

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}


# 제목, 본문 담을 리스트
contents = []
date_list = []

def crawl_edaily(start, end, endpage):

    edaily_url = f'https://www.edaily.co.kr/search/news/?source=total&keyword=금리&include=&exclude=&jname=&start={start}&end={end}&sort=latest&date=pick&exact=false&page='
    # 시작 페이지 설정
    page_number = 1

    while True:
        # 현재 페이지 URL 생성
        current_url = edaily_url + str(page_number)
        
        # 페이지 요청
        response = requests.get(current_url, headers=headers)
        soup = bs(response.text, 'html.parser')

        news_ls = soup.select('.newsbox_texts')

        # 뉴스기사 제목, 내용 크롤링
        for content in news_ls:
            contents.append(content.text.strip())

        # 기사 날짜 크롤링
        for i in soup.select('.author_category'):
            date_list.append(i.text.split()[0])

        if page_number == endpage:
            break
        page_number += 1
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}

# 제목, 본문, 날짜, URL을 담을 리스트
contents = []
date_list = []
url_list = []

def crawl_edaily(start, end, endpage):

    edaily_url = f'https://www.edaily.co.kr/search/news/?source=total&keyword=금리&include=&exclude=&jname=&start={start}&end={end}&sort=latest&date=pick&exact=false&page='
    # 시작 페이지 설정
    page_number = 1

    while True:
        # 현재 페이지 URL 생성
        current_url = edaily_url + str(page_number)
        
        # 페이지 요청
        response = requests.get(current_url, headers=headers)
        soup = bs(response.text, 'html.parser')

        # 뉴스기사 제목, 내용 크롤링
        for news_item in soup.select('.newsbox_04 .newsbox_texts'):
            # 뉴스 텍스트 크롤링
            contents.append(news_item.text.strip())
            
            # 기사 URL 크롤링
            link_tag = news_item.find_previous('a')
            if link_tag:
                url = 'https://www.edaily.co.kr' + link_tag['href']
                url_list.append(url)
            else:
                url_list.append(None)

        # 기사 날짜 크롤링
        for i in soup.select('.author_category'):
            date_list.append(i.text.split()[0])

        if page_number == endpage:
            break
        page_number += 1
```

```python
def save_to_csv(contents, date_list, filename='output.csv'):
    # 데이터프레임 생성
    data = {
        "DATE": date_list,
        "CONTENTS": contents,
        "URL": url_list
    }
    df = pd.DataFrame(data)
    
    # 날짜순으로 정렬
    df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')  # 날짜 형식으로 변환
    df = df.sort_values(by='DATE')

    # 번호 추가
    df.insert(0, 'No.', range(1, len(df) + 1))
    
    # CSV 파일로 저장
    df.to_csv(filename, index=False, encoding='utf-8-sig')
```

```python
all_contents = []
all_dates = []
all_url = []

# 여러 기간별
def run_all_crawls():
    periods = [
        (20140811, 20150131, 259),
        (20150201, 20150731, 322),
        (20150801, 20160131, 344),
        (20160201, 20160731, 257),
        (20160801, 20170131, 261),
        (20170201, 20170731, 210),
        (20170801, 20180131, 178),
        (20180201, 20180731, 173),
        (20180801, 20190131, 180),
        (20190201, 20190731, 136),
        (20190801, 20200131, 24),
        (20200201, 20200731, 133),
        (20200801, 20210131, 129),
        (20210201, 20210731, 197),
        (20210801, 20220131, 208),
        (20220201, 20220731, 308),
        (20220801, 20230131, 390),
        (20230201, 20230731, 314),
        (20230801, 20240131, 395),
        (20240201, 20240811, 399)
    ]

    for start, end, endpage in periods:
        crawl_edaily(start, end, endpage)
        all_contents.extend(contents)
        all_dates.extend(date_list)
        all_url.extend(url_list)
        contents.clear()
        date_list.clear()
        url_list.clear()

# 크롤링 실행
run_all_crawls()

# 데이터프레임으로 변환 및 정렬
def save_to_csv_from_combined(contents, date_list, url_list, filename='output.csv'):
    # 데이터프레임 생성
    data = {
        "DATE": date_list,
        "CONTENTS": contents,
        "URL": url_list
    }
    df = pd.DataFrame(data)
    
    # 날짜순으로 정렬
    df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')  # 날짜 형식으로 변환
    df = df.sort_values(by='DATE')

    # 번호 추가
    df.insert(0, 'No.', range(1, len(df) + 1))
    
    # CSV 파일 저장
    df.to_csv(filename, index=False, encoding='utf-8-sig')

# 모든 크롤링 데이터 저장
save_to_csv_from_combined(all_contents, all_dates, all_url, 'combined_edaily_2.csv')
```
